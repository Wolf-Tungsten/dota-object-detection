```python
import tensorflow as tf
import cv2
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
def load_image( path ):
    # 读取图片成numpy数组
    image = cv2.imread(path)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    label_file_path = path.split('.')[0] + '.txt'
    canvas = image.copy()
    with open(label_file_path, 'r') as label_file:
        for line in label_file.readlines():
            line = line.split(' ')
            point1 = (int(line[1]), int(line[2]))
            point2 = (int(line[3]), int(line[4]))
            cv2.rectangle(canvas,point1,point2,(0, 255, 0),3)
    plt.imshow(image)
    plt.imshow(canvas)
    print(image.shape)
    return image
```
